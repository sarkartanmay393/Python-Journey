```python
import pandas
from geopy.geocoders import ArcGIS

arc = ArcGIS()

data = pandas.read_excel("p7-related.xlsx")
data
```

```python
data["Address"] = data["Address"] + ", " + data["City"] + ", " + data["State"]
# data
```

```python
data["Coordinates"] = data["Address"]
# data
```

```python
data["Coordinates"] = data["Address"].apply(arc.geocode)
# data
```

```python
data["Latitude"] = data["Coordinates"].apply(lambda x : x.latitude if x!=None else None)
# data
```

```python
data["Longitude"] = data["Coordinates"].apply(lambda x : x.longitude if x!=None else None)
# data
```

```python
data.drop(columns='Coordinates')
```
